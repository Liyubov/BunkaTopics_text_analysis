```python
def preprocess_data(data):
    processed_data = []
    for record in data:
        authors = [author['name'] for author in json.loads(record['author'])]
        record['authors'] = authors
        del record['author']
        processed_data.append(record)
    return processed_data

processed_data = preprocess_data(data)
```
